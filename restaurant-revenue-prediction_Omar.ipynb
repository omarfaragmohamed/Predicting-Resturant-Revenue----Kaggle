```python
#Handling the date and dropping the Open Date column   -- train Data
train_data['Open Date'] = pd.to_datetime(train_data['Open Date'])
train_data['Month'] = [x.month for x in train_data['Open Date']]
train_data['year'] = [x.year for x in train_data['Open Date']]
train_data.drop(['Open Date'],axis=1,inplace=True)
```

```python
#Handling the date and dropping the Open Date column  -- Test Data
test_data['Open Date'] = pd.to_datetime(test_data['Open Date'])
test_data['Month'] = [x.month for x in test_data['Open Date']]
test_data['year'] = [x.year for x in test_data['Open Date']]
test_data.drop(['Open Date'],axis=1,inplace=True)
```

##### We will use fit and transform on test data as it has the max number of categories

```python
encoder = LabelEncoder()
train_data['City Group'] = encoder.fit_transform(train_data['City Group'])
test_data['City Group'] = encoder.transform(test_data['City Group'])

# Convert the method of fitting and transforming
test_data['Type'] = encoder.fit_transform(test_data['Type'])
train_data['Type'] = encoder.transform(train_data['Type'])
```

```python
#Dummies -- Train data
train_data = pd.get_dummies(train_data, columns=['Month','Type','City Group'], prefix = ['Month','Type','City Group'])

#Dummies -- Train data
test_data = pd.get_dummies(test_data, columns=['Month','Type','City Group'], prefix = ['Month','Type','City Group'])
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
# Seperating the data
y = train_data['revenue']
x = train_data.copy()
x.drop(['revenue'],axis=1,inplace=True)

x_test = test_data
```

```python
# Adjust the shapes
x ['Type_3'] = 0
print (x .shape)
print (y.shape)
print (x_test.shape)
```

```python
# Splitting the data
x_train, x_val, y_train, y_val = train_test_split(x,y, test_size = 0.25,shuffle= False, random_state = 42)
```

### Modeling the Data

#### We will use grid search to apply some of models and compare between their performances

```python
# Setting Parameters
regressors = {
    'Linear Regression' : LinearRegression(),
    'Decision Tree' : DecisionTreeRegressor(),
    'Random Forest' : RandomForestRegressor(),
    'AdaBoostRegressor': AdaBoostRegressor(),
    'K-nearest Neighbors' : KNeighborsRegressor(),
    'GradientBoostingRegressor' : GradientBoostingRegressor(),
    'Support Vector Machines' : SVR(),
    'XGBoost' : XGBRegressor()
}
results=pd.DataFrame(columns=['MAE','MSE','R2-score'])

for method,func in regressors.items():
    func.fit(x_train,y_train)
    y_pred = func.predict(x_val)
    results.loc[method]= [mean_absolute_error(y_val,y_pred),
                          mean_squared_error(y_val,y_pred),
                          r2_score(y_val,y_pred)
                         ]
results
```

```python
#RandomForestRegressor

parameters = {
              'max_depth':  list(range(6, 30, 10)),
              'max_leaf_nodes': list(range(50, 500, 100)),
              'n_estimators': list(range(50, 1001, 150))
            }

from sklearn.model_selection import GridSearchCV

gsrch = GridSearchCV(estimator=RandomForestRegressor(),
                       param_grid = parameters, 
                       scoring='neg_mean_squared_error',
                       n_jobs=4,cv=5,verbose=7)

gsrch.fit(x,y)
```

```python
print(gsrch.best_params_.get('n_estimators'))
print(gsrch.best_params_.get('max_leaf_nodes'))
print(gsrch.best_params_.get('max_depth'))
print (gsrch.best_score_)
```

```python
RF_model = RandomForestRegressor(
                         max_depth = gsrch.best_params_.get('max_depth'),
                           max_leaf_nodes = gsrch.best_params_.get('max_leaf_nodes'),
    n_estimators = gsrch.best_params_.get('n_estimators'),random_state=42, n_jobs=4)
RF_model.fit(x,y)
```

```python
y_prd_val_RF= RF_model.predict(x)
y_prd_RF_final = RF_model.predict(x_test)
```

```python
R2=r2_score(y,y_prd_val_RF)
R2
```

```python
MSE = np.square(np.subtract(y,y_prd_val_RF)).mean() 
MSE
```

```python
final_submission = pd.DataFrame(columns=['Id','Prediction'])
final_submission['Id'] = test_data_row['Id']
final_submission['Prediction'] = y_prd_RF_final
final_submission.to_csv('submission_RF_final.csv',index=False)
```
